```python
import pandas as pd
df = pd.read_csv('./titanic/train.csv')
```

```python
df["Name"].sample(20)
```

```python
def get_title(name):
    if '.' in name:
        return name.split(',')[1].split('.')[0].strip()
    else:
        return "No title in name"
```

```python
titles = set([x for x in df.Name.map(lambda x: get_title(x))])
```

```python
print(titles)
```

```python
def short_titles(x):
    title = x["Title"]
    if title in ['Capt', 'Col', 'Major']:
        return 'Oficcer'
    elif title in ["Jonkheer", "Don", "The Countees", "Dona", "Lady", "Sir"]:
        return "Royalty"
    elif title == "Mme":
        return "Mrs"
    elif title in ["Mlle", "Ms"]:
        return "Miss"
    else:
        return title
```

```python
df["Title"] = df["Name"].map(lambda x : get_title(x))
```

```python
df["Title"] = df.apply(short_titles, axis=1)
```

```python
print(df.Title.value_counts())
```

```python
df.drop('Cabin', axis=1, inplace=True)
```
